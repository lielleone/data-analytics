## 5. Changing values on a slice

Looking at only part of a dataframe is called **slicing**, or a **view**

But this is still the ORIGINAL dataframe

Here, we think we are creating a copy of the dataframe. But actually, we are just naming a slice of the dataframe

```python
united_2015_df = strike_df[(strike_df.Operator == 'UNITED AIRLINES') & (strike_df['Incident Year'] == 2015)]
united_2015_df.head(1)
```

```python
united_2015_df['Airport'][124866]
```

Let's change it to TEL AVIV\
We are warned that we are trying to change both our slice, and the original strike_df dataframe.\
Usually, if we are trying to change data is a slice, it mean we don't want to change data in the original dataframe

```python
united_2015_df['Airport'][124866] = "TEL AVIV"
```

The way to avoid the warning is to use `loc` or `iloc`.

```python
united_2015_df.loc[124866,'Airport'] = "LOD"
united_2015_df['Airport'][124866]
```

```python
strike_df['Airport'][124866]
```

This works fine, no warnings. This won't change the original dataframe (which is a good things since if you planned to change it you would work on it and not on a slice)\
This will be the only way to do it from Pandas 3.0 onward\
[more explanations](https://pandas.pydata.org/pandas-docs/stable/user_guide/copy_on_write.html#copy-on-write)

## 6. `loc` and `iloc` 

One way to do that is iloc. 

`.iloc` - selects subsets of rows and columns by integer location only

```python
#strike_df.iloc[[70000]]  #first row as a series
#strike_df.iloc[[5,50,50000]]  #first row as a dataframe
#strike_df.iloc[-1] #last row as a series
strike_df.iloc[1:5,] #last row as a dataframe
```

The : operator 

 - when used alone it means "everything"

- also used to indicate a ***slice*** of values


```python
strike_df.iloc[2:4] # second and third row
#strike_df.iloc[[-1,2,22]] #a few specific rows

# Columns:
strike_df.iloc[:,0] # first column of data frame  
strike_df.iloc[:,[1,2]] # second column of data frame  
#strike_df.iloc[:,-1] # last column of data frame

#Rows and columns
#strike_df.iloc[0:5] # first five rows of dataframe
strike_df.iloc[4:6, 0:2] # first two columns of data frame with all rows
#strike_df.iloc[[0,3,6,24], [0,5,6]] # 1st, 4th, 7th, 25th row + 1st 6th 7th columns.
```

What if I want to select the 'daily_vaccinations' column, but I don't remember the column number?

Use `.loc`

`.loc` - selects subsets of rows and columns by label only. Allowed inputs are:

- A single label, e.g. 5 or 'a', (note that 5 is interpreted as a label of the index, and never as an integer position along the index).

- A list or array of labels, e.g. ['a', 'b', 'c'].

- A slice object with labels, e.g. 'a':'f'.

```python
strike_df.loc[2:4,['Airport','Incident Year']]
```

Lets add the species name

```python
strike_df.loc[0:3,['Airport','Incident Year','Species Name']]
```

Semantics are similar to iloc. But note:

- `iloc` excludes the last element.  `df.iloc[0:1000]` will return entries 0...999
- `loc`, includes the last element.  `df.loc[0:1000]` will return entries 0...1000
