```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
from datetime import date, datetime, timedelta
```

```python
url = 'https://raw.githubusercontent.com/barakelitzur/final_project/main/Airplane_Crashes_and_Fatalities_Since_1908_20190820105639.csv'
Crashes_df = pd.read_csv(url)
print("The data has", Crashes_df.shape[0], "rows and", Crashes_df.shape[1], "columns.")
```

```python
Crashes_df.columns =Crashes_df.columns.str.replace('#', 'number')
Crashes_df['Date'] = pd.to_datetime(Crashes_df['Date'])
Crashes_df['year'] = pd.DatetimeIndex(Crashes_df['Date']).year
Crashes_df['month'] = pd.DatetimeIndex(Crashes_df['Date']).month
Crashes_df['time'] = pd.DatetimeIndex(Crashes_df['Date']).month
```

# התרסקויות לפי שנים

```python
Crashes_by_year_df=Crashes_df.groupby("year")[["Date"]].count().reset_index()
year_crashes = pd.DataFrame({'year': Crashes_by_year_df["year"], 'crashes': Crashes_by_year_df["Date"]})
pd.pivot_table(year_crashes,index = ['year'], values = ['crashes']).plot(figsize=(15,7),xticks=(1910,1920,1930,1940,1950,1960,1970,1980,1990,2000,2010,2020))
plt.title("Crashes by years",fontsize=20 , color = "blue")
```

ניתן לראות בגרף עליה חדה בהתרסקויות המטוסים בין השנים 1940 עד 2000 עקב השינויים הדרסטיים במהירויות המטוסים שהתפתחו עם השנים 
וירידה לקראת שנת 2000 שלהערכתנו נובעת מההתפתחות הטכנולוגית והיציבות בענף התעופה.

# עשרת סוגי המטוסים שהתרסקו הכי הרבה פעמים

```python
Crashes_by_ac=Crashes_df.groupby("AC Type")[["Fatalities"]].count().reset_index()
Crashes_by_ac.sort_values("Fatalities", ascending = False)
fig, axes = plt.subplots(figsize=(10,5))
sns.countplot(y='AC Type', data=Crashes_df, order=Crashes_df['AC Type'].value_counts().index, palette='Set3')
sns.despine(bottom=False, left=True)
axes.set_ylim([9, 0])
axes.set_title('Top 10 planes crashed by type', size=20, color = "blue")
axes.set(xlabel='Number of crashes', ylabel='AC Type')
plt.tight_layout()
```

# אחוז ניצולים לפי סוג מטוס

```python
Crashes_df['Survivors'] = Crashes_df['Fatalities'] / Crashes_df['Aboard']
Crashes_df['Survivors'] =(1-Crashes_df['Survivors'])*100 
Crashes_df=Crashes_df.dropna (subset = ["Survivors"])
Crashes_df.loc[Crashes_df['Survivors'].isnull() == True]

Crashes_big_df = Crashes_df[Crashes_df["Aboard"]>=100]

Crashes_by_ac=Crashes_big_df.groupby("AC Type")[["Fatalities","Survivors"]].mean().reset_index()
Crashes_by_ac.rename(columns={'Fatalities':'Number of crashes'},inplace=True)
Crashes_by_ac.sort_values("Number of crashes", ascending = False)
graph = Crashes_by_ac[Crashes_by_ac["Number of crashes"]==1].sort_values("Survivors").tail()
graph
#Crashes_by_ac
```

במידה וחייבים לטוס במטוס שהתרסק  בעבר, נעדיף לטוס במטוס שבו היו מינימום התרסקויות ואחוז ניצולים גבוה. נשער שמטוס זה הוא הבטוח ביותר מכיוון שסוג זה גם לא התרסק יותר מפעם אחת וגם אחוז הניצולים בו הוא הגבוה ביותר.

# כמות ניצולים ביחס לכמות הצוות במטוס

```python
Crashes_df=Crashes_df.dropna (subset = ["Fatalities Crew","Fatalities Passangers","Fatalities","Aboard Crew","Aboard Passangers","Aboard"]).copy()
Crashes_big_df = Crashes_df[Crashes_df["Aboard"]>=100].copy()
Crashes_big_df['survive'] = Crashes_big_df['Aboard']-Crashes_big_df['Fatalities']
Crashes_big_df['survive'] =(Crashes_big_df['survive'])*0.01 
Crashes_big_df["crew_per"] = Crashes_big_df["Aboard Crew"]/Crashes_big_df['Aboard']

fig, axes = plt.subplots(figsize=(14, 5))
plt.title("Lots of survivors in relation to the amount of staff",fontsize=20, color = "blue")
sns.lineplot(x="crew_per", y='survive', data=Crashes_big_df)
```

נסינו לקשר בין כמות הניצולים לכמות הצוות, גלינו שדווקא במטוסים בהם אחוז אנשי הצוות היה נמוך הייתה כמות גדולה יותר של ניצולים.
