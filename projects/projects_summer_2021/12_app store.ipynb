```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt
import seaborn as sns
```

# Data Cleaning Tasks to be carried out as follows.
### 1. Rename Columns to one word.
### 2. Fillna with 0.
### 3. Remove suffixes like '+' in Installs and '$' in price .
### 4. Change format of columns such as Last Updated , Installs,  Size,  Reviews ,Rating.


```python
data=pd.read_csv("googleplaystore.csv")
data.isnull().sum()
```

```python
data.isnull().sum()
#Checking null values in the data.

```

```python
data['Rating'].fillna('0',inplace = True)
# put 0 to data that dont have value
```

```python
data.isnull().sum()
```

```python
data.dtypes
```

```python
#data['Size'] = data['Size'].str.replace('M','').astype('float')
indexNames = data[ data['Size'] == 'Varies with device' ].index
data.drop(indexNames , inplace=True)
data.head()
```

```python
data['Size']=data['Size'].str[:-1]
data['Size'] = data['Size'].astype(float)
data.head()
#convert to float
```

```python
#Delete ',' and '+' from de column
data['Installs'] = data['Installs'].replace({',': ''}, regex=True)
data['Installs']=data['Installs'].str[:-1]

#convert to numeric
data['Installs']= pd.to_numeric(data['Installs'])

data.head()
```

```python
#we change that to date type
data['Last Updated'] = pd.to_datetime(data['Last Updated'])
```

```python
data['Rating'] = data['Rating'].astype('float')
#convert to float
```

```python
data.dtypes
```

```python
data.isnull().sum()
```

```python
data.Rating.max()
#Rating need to be beetween 0 to 5 .
```

```python
data.Price.value_counts()
```

```python
data['Price'] = data['Price'].str.replace('$', '')
data.Price.value_counts()
#Delete '$'' from column Price
```

```python
data['Price']=pd.to_numeric(data['Price'])

#convert to numeric
```

```python
data.dtypes
```

```python
Category = data.groupby(['Category'])['App'].count().reset_index().rename(columns={'App':'Count of Apps'})
Category = Category.sort_values(by=['Count of Apps'], ascending=False)
Category.set_index('Category', inplace= True)
Category.head()
```

```python
Reviews_Cat2= data[data['Category'].isin(data.Category)]
```

```python
fig5, ax = plt.subplots(figsize=(20,3))
sns.countplot(x=Reviews_Cat2['Category'],order=Reviews_Cat2['Category'].value_counts().index,hue = Reviews_Cat2['Type'], saturation=1,edgecolor=(0,0,0),linewidth=2)
plt.xlabel('\n Categories', fontsize = 10, fontweight = 'bold')
plt.ylabel('Count of Apps', fontsize = 10, fontweight = 'bold')
plt.title('Apps per Category for Highest Review Categories', fontsize = 15, fontweight = 'bold')
plt.savefig('Apps per Category for Highest Review Categories.png')
plt.show()
```

```python
data_free=data[data['Type']=="Free"]
data_free.tail()
```

```python
data_cut2 = data_free.groupby(['Category'])[['Reviews','Installs']].sum().sort_values('Installs', ascending = False)
data_cut2.head()
```

```python
plt.figure(figsize=(7,4)) 
sns.heatmap(data_cut2.corr(),annot=True,cmap='Greens') #draws  heatmap with input as the correlation matrix calculted by(iris.corr())
plt.show()
```

```python
sorted_data=data_free.sort_values(by="Installs",ascending=False)
plt.figure(figsize=(15,10))
sns.barplot(x=sorted_data['Installs'], y=sorted_data['Reviews'])
plt.xticks(rotation=80)
plt.xlabel("Installs")
plt.ylabel("Reviews(M)")
plt.title("Install per Reviews")
plt.show()
```
