### Correlations

```python
corrs = sr_df.corr(method = 'kendall')

plt.figure(figsize=(15,10))  #figure size
sns.heatmap(corrs, cmap='coolwarm', center=0, annot = True);
```

We can tell that there is a diffrent connection between the Urban and Rural residence and the Total of dropout. 


```python
D = sr.pivot_table(['Rural_Residence','Urban_Residence'],'Total') 
D['Total'] = sr['Total']
D.head()
```

```python
sr['Development Regions'].value_counts()
```

```python
fig, ax = plt.subplots(1,2,figsize=(10, 5))
sns.regplot(x='Total',y='Rural_Residence', data=sr, scatter_kws={"color": "Pink"}, line_kws={"color": "Green"}, ax = ax[0])
sns.regplot(x='Total',y='Urban_Residence', data=sr, scatter_kws={"color": "Pink"}, line_kws={"color": "Green"}, ax = ax[1])
plt.subplots_adjust(wspace = 0.3)
```


>##### Observation 1:
>The relationship between the total percentage of dropouts and the percentage of urban dropouts is very close. 
>But the connection between total and rural is almost perfect.
>We conclude that the Rural Residence has much more influence on the total percentage of dropouts.
